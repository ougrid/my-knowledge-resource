```python
master_df = pd.read_csv('/content/dataset/image_class.csv')
master_df = master_df.astype(str)
master_df 
```

```python
src_path = '/content/dataset/im_train/'
des_path = '/content/image_dataset/train/'
for i in range(len(master_df)):
# for i in range(3):
  src_file = src_path + master_df['filename'][i]
  des_file = des_path + master_df['classes'][i] + '/' + master_df['filename'][i]
  shutil.copy(src_file, des_file)
```

```python
input_folder = '/content/image_dataset/train'  
output_folder= '/content/image_dataset/output'
splitfolders.ratio(input_folder, output= output_folder, seed=1337, ratio = (0.8, 0, 0.2))
```

```python
train_dataset, _, test_dataset = ImageDataset.from_folders(output_folder)
print(train_dataset)
```

```python
ImagePredictor.list_models()
```

```python
predictor = ImagePredictor()
predictor.fit(train_dataset, hyperparameters={'epochs': 1}) 
```

```python
fit_result = predictor.fit_summary()
print('Top-1 train acc: %.3f, val acc: %.3f' %(fit_result['train_acc'], fit_result['valid_acc']))
```

```python
test_dataset.head()
```

```python
bulk_result = predictor.predict(test_dataset)
print(bulk_result)
```
