```python
training_data = dataset_train.iloc[:,5:6]
```

```python
plt.figure(figsize=(10,5))
plt.plot(training_data, label='Training Data')
plt.ylabel("Stock Price")
plt.title("ABC Mockup Stock")
plt.xlabel("Time") # 2k days
plt.show()
```

```python
# ML can work extremely well when data is in the range of 0 to 1, so we need to scale the data with MinMaxScaler
nm = MinMaxScaler(feature_range=(0, 1))
training_data_scaled = nm.fit_transform(training_data) # nm stands for MinMax
```

```python
training_data_scaled
```

```python
training_data_scaled.shape # (rows, columns)
training_data_scaled.shape[0]
```

```python
plt.figure(figsize=(10,5))
plt.plot(training_data_scaled, label='Training Data', color='pink') # Scaled data is in the range of 0 to 1
plt.ylabel("Stock Price")
plt.title("ABC Mockup Stock")
plt.xlabel("Time") # 2k days
plt.show()
```

```python
x_train = training_data_scaled[59:training_data_scaled.shape[0]-365-1] # The first 60 days and the last 365 days are not used, also lagged for 1 day
y_train = training_data_scaled[60:training_data_scaled.shape[0]-365]

x_test = training_data_scaled[training_data_scaled.shape[0]-365-1:-1]
y_test = training_data_scaled[training_data_scaled.shape[0]-365:]

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], 1, 1))
x_train.shape
```

```python
import re
# LSTM model solve the problem of vanishing gradient and overfitting (model being forgetful)
# Nowadays we use GRU, or Transformer which perform better

regressor = Sequential()
regressor.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
regressor.add(Dropout(0.2)) # To avoid overfitting by dropping some neurons | Normally we use 0.2-0.3 | Empirical | Use dropout only if your model is so genral, not specific
regressor.add(LSTM(units = 50,return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50,return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units = 1)) # 1 output (If there are 3 classes, then there will be 3 outputs, 3 dense layers)
```

```python
regressor.summary()
# Normally No. of parameters = (No. of Features + 1) * No. of Neurons in 1st Layer | Macbook can handle 2-3M Params easily in 1 hour training

# regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# regressor.fit(x_train, y_train, epochs = 100, batch_size = 32)
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
# Adam is currently the best optimizer
# There are many other loss functions like MSE, MAE, Precision, Recall, F1 --> Choose the appropriate one to your output
# MSE
```

```python
regressor.fit(x_train, y_train, epochs = 50, batch_size = 32) # batch_size = 32 --> choose whatever powers of 2
# Around Epoch 20/50, models are already fit to the data, so actually we can stop training at that point.
```

```python
x_test = np.reshape(x_test, (x_test.shape[0], 1, 1))
```

```python
# regressor.predict(x_test)
pred = regressor.predict(x_test)
```

```python
pred_inverse = nm.inverse_transform(pred)
y_test_inverse = nm.inverse_transform(y_test)
```

```python
pred_inverse
```

```python
y_test_inverse # gold price
```

```python
plt.figure(figsize=(15,8))
plt.plot(pred_inverse, label='Predicted Stock Price', color='green')
plt.plot(y_test_inverse, label='Actual Stock Price', color='blue')
plt.ylabel("Stock Price")
plt.title("ABC Mockup Stock")
plt.xlabel("Time") # 2k days
plt.show()
```

```python
rmse = mean_squared_error(y_test_inverse, pred_inverse, squared=False)
```

```python
print("rmse: ", rmse)
# The lower the better, rmse:  4.0585092291202 = the average difference between predicted and actual stock price
```
